```python
# Import necessary libraries
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Load dataset
# Assuming that 'data.csv' is your dataset and it's in the same directory
data = pd.read_csv("/home/dev/Documents/Input/ALL CDN Data.csv")
data = data[data["Community"] == "port credit"]
data = data.dropna()
```

```python
data.head()
```

```python
columns = [
    "lat",
    "lng",
    "calculated_Approx_Age",
    "latest_tax",
    "Washrooms",
    "Bedrooms",
    "Bedrooms_",
    "Adjusted_Total_Parking_Spaces",
    "Parking_Category",
    "Modified_Balcony",
    "Approx_Square_Footage",
    "FLOOR_NUMBER",
    "Current_Price",
    "Address",
    "Sold_Month",
]
categorical_columns = [
    "Washrooms",
    "Bedrooms",
    "Adjusted_Total_Parking_Spaces",
    "Parking_Category",
    "Modified_Balcony",
    "FLOOR_NUMBER",
    "calculated_Approx_Age",
    "Approx_Square_Footage",
    # "Address",
    "Bedrooms_",
]
numerical_columns = [
    "latest_tax",
]

data = data[columns]
```

```python
data.head()
```

```python
for column in categorical_columns:
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])
    
```

```python
 #Preprocessing
# Assuming that the last column is 'price' and the others are features
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Split the dataset into the training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Initialising the MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(32, 32), activation='relu', solver='adam', max_iter=500)

# Fitting the MLPRegressor to the Training set
model.fit(X_train, y_train)

# Predicting the Test set results
y_pred = model.predict(X_test)
```
